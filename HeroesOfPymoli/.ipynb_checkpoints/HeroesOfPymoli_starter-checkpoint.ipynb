```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players

```python
#Count number of Total Players
Total_Players = purchase_data["SN"].nunique()
player_count = pd.DataFrame({"Total Players":[Total_Players]})
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
#Count number of unique Items
Unique_Item = purchase_data["Item ID"].nunique()
Average_Price=round(purchase_data["Price"].mean(),2)
Purchase_Numbers = purchase_data["Item ID"].count()
Total_Revenue = purchase_data["Price"].sum()

#Create a Data Frame for Purchasing Analysis
purchasing_analysis = pd.DataFrame({"Number of Unique Items":[Unique_Item],
                                    "Average Price":[Average_Price],
                                   "Number of Purchase":[Purchase_Numbers],
                                    "Total Revenue":[Total_Revenue]})

#Format the Prie and revnue with $sign
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${:.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:.2f}".format)

purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Group by Gender to show the information by gender
Gender = purchase_data.groupby(["Gender"])
Gender_Count = Gender["SN"].nunique()
Percentage = (Gender_Count/Total_Players) * 100

#Use map to format the data with the appropriate formating
Gender_Summary = pd.DataFrame({"Gender Count":Gender_Count,"Percentage":Percentage})
Gender_Summary["Percentage"] = Gender_Summary["Percentage"].map("{:.2f}%".format)

Gender_Summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Purchase = purchase_data.groupby(["Gender"])
Purchase_Count = Purchase["Purchase ID"].count()
Average_Price = Purchase["Price"].mean()
Total_Value = Purchase["Price"].sum()

Average_Person = Total_Value/Gender_Count

Gender_Summary_two = pd.DataFrame({"Purchase Count":Purchase_Count,"Average Purchase Price":Average_Price,
                                "Total Purchase Value":Total_Value,"Avg Total Purchase per Person":Average_Person})

#Use map to format the data with the appropriate formating
Gender_Summary_two["Average Purchase Price"] = Gender_Summary_two["Average Purchase Price"].map("${:.2f}".format)
Gender_Summary_two["Total Purchase Value"] = Gender_Summary_two["Total Purchase Value"].map("${:.2f}".format)
Gender_Summary_two["Avg Total Purchase per Person"] = Gender_Summary_two["Avg Total Purchase per Person"].map("${:.2f}".format)
Gender_Summary_two 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins for the different age group
bins = [0,9,14,19,24,29,34,39,200]
labels= ["<10", "10-14" , "15-19" , "20-24" , "25-29" , "30-34","35-39", "40+"]
age_df = purchase_data.copy()
age_df["Age Ranges"] = pd.cut(age_df["Age"], bins, labels=labels, include_lowest=True)
age_group = age_df.groupby("Age Ranges")

total_count = age_group["SN"].nunique()
percentage = total_count/total_count.sum()*100
demo_df = pd.DataFrame({"Total Count":total_count, "Percentage":percentage}) 

#Use map to format the data with the appropriate formating
demo_df["Percentage"] = demo_df["Percentage"].map("{:.2f}%".format)

demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Create bins for the different age group
bins = [0,9,14,19,24,29,34,39,200]
labels= ["<10", "10-14" , "15-19" , "20-24" , "25-29" , "30-34","35-39", "40+"]
age_df = purchase_data.copy()
age_df["Age Ranges"] = pd.cut(age_df["Age"], bins, labels=labels, include_lowest=True)
age_group = age_df.groupby("Age Ranges")

purchase_count = age_group["Purchase ID"].count()
average_price = age_group["Price"].mean()
total_purchase = age_group["Price"].sum()
average_person = total_purchase/total_count

demo_df = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":average_price, 
                        "Total Purchase Value":total_purchase, "Avg. Total Purchase per person":average_person}) 
#Use map to format the data with the appropriate formating
demo_df["Average Purchase Price"] = demo_df["Average Purchase Price"].map("${:.2f}".format)
demo_df["Total Purchase Value"] = demo_df["Total Purchase Value"].map("${:.2f}".format)
demo_df["Avg. Total Purchase per person"] = demo_df["Avg. Total Purchase per person"].map("${:.2f}".format)

demo_df
```
